```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
from IPython.display import HTML
requests.packages.urllib3.disable_warnings()
import warnings
warnings.filterwarnings('ignore')
page_num = 1
while True:
  url = f"https://clevershop.mk/product-category/mobilni-laptopi-i-tableti/page/{page_num}/"
  snapshot = requests.get(url)
  if snapshot.status_code != 200:
    print("No more pages found")
    break
  soup = BeautifulSoup(snapshot.text,'html.parser')
  titles = []
  reg = []
  diss = []
  purl = []
  atcurl = []
  prices = soup.select('.products .price')
  for i in prices:
    ceni = i.select('.woocommerce-Price-amount')
    if len(ceni) > 0:
      if len(ceni) == 1:
        disscount = None
      else: disscount = ceni[1].text.strip()
      reg.append(ceni[0].text.strip())
      diss.append(disscount)
    else:
      reg.append(None)
      diss.append(None)

  products = soup.select('.products .wd-entities-title')
  for i in products:
    titles.append(i.text.strip())

  karti = soup.select('.wd-entities-title a')
  for i in karti:
    purl.append(i['href'])
  
  carts = soup.select('.wd-add-btn .add_to_cart_button')
  for i in carts:
    atcurl.append(i['href'])
  
  data={"Tiltles":titles,
     "Regular price":reg,
     "Disscount price":diss,
     "Product url":purl,
     "Add to cart url":atcurl}
  
  df = pd.DataFrame(data)
  print(df)
  page_num= page_num + 1
```
